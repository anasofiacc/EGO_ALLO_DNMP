```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

##### Read the data file

```python
latencies=(pd.read_excel('LATENCIES CATCH SESSIONS.xlsx', index_col=None))
latencies.head()
```

##### Calculate the latency to express choice (latency t1 + latency t2)

```python
latencies['latency_to_choice']=latencies['latency_t1'] + latencies['latency_t2']
```

##### Does latency to express choice increase in catch trials?

```python
#Subset the data
subset=latencies[(latencies['trial']>=-1)]

PROPS = {
    'boxprops':{'facecolor':'none'}}
plt.figure(figsize=(6,4))
g=sns.boxplot(
    data=subset, x='trial', 
    y='latency_to_choice', color='black', **PROPS)

g.set(xlabel='trial', ylabel='Latency to choice expression (s)', 
      xticklabels=['PRE-CATCH','CATCH'], xlim=[-.5,1.5], ylim=[0,30])

sns.despine()
```

```python
#Subset the data
subset=latencies[(latencies['trial']>=-1)]

colors=['orange', 'indigo', 'deepskyblue', 'crimson']
sns.set(context='talk')

g=sns.catplot(kind='point', col='catch_nr', sharey=True, col_wrap=2,
              data=subset, x='trial', y='latency_to_choice', ci='sd', estimator=np.mean,
              join=True, hue='catch_nr', palette=colors,
              height=3, aspect=1.2)

ax=g.axes.flatten()
for i in range(4):
    ax[i].set_title('Catch number {}'.format(i+1))
    # Collect data by catch number
    data=subset[subset['catch_nr']==i+1]
    data['trial']+=1
    sns.lineplot(data=data, x='trial', y='latency_to_choice', hue='rat', legend=False, ax=ax[i],
                linewidth=.5, palette=12*[colors[i]])
    
g.set(xlabel='trials', ylabel='Latency (s)\n', 
      xticklabels=['PRE-CATCH','CATCH'], xlim=[-.3,1.3])
sns.despine()
```

##### How do the latencies between catch trials compare?

```python
subset=latencies[latencies['trial']==0]
```

```python
colors=['orange', 'indigo', 'deepskyblue', 'crimson']
sns.set(context='talk')
g=sns.pointplot(
              data=subset, x='catch_nr', y='latency_to_choice', 
              ci='sd', estimator=np.mean, join=True, color='black')
subset['catch_nr']=subset['catch_nr']-1
sns.lineplot(data=subset, x='catch_nr', y='latency_to_choice', 
             hue='rat', legend=False, ci=None,
             linewidth=.5, palette=12*['black'])
 
g.set(ylabel='latency to choice expression (s)', xlabel='Catch number')
sns.despine()
```

##### Is there a difference between the latencies from non-matches to location vs. non-matches to action catch trials?

```python
catch_info=pd.read_excel('ALLOEGOPROTOCOL_OUTCOMES.xlsx', index_col=None).iloc[:,0:6]
cols=catch_info.columns.to_list()
lowered_cols= [col.lower() for col in cols]
catch_info.columns=lowered_cols
```

```python
catch_latencies=latencies[latencies['trial']==0].reset_index(drop=True)
```

```python
cols=['latency_t1', 'latency_t2', 'latency_to_choice']
catch_info[cols]=catch_latencies[cols]
catch_info.head()
```

```python
def plot_latencies_by_outcome(col, title):
    plt.figure(figsize=(6,4))
    g=sns.boxplot(data=catch_info, y=col, x='outcome',
           palette=['mediumseagreen', 'royalblue'])
    g.set(xlim=[-.6,1.6], ylabel='Latency to choice point', xlabel='Non match', 
      xticklabels=['Action', 'Location'], title=title)
    sns.despine()
```

```python
plot_latencies_by_outcome('latency_to_choice', "Latency to choice expression in catch trials" )
```

```python
plot_latencies_by_outcome('latency_t1', 'Latency to choice point in catch trials')
```
