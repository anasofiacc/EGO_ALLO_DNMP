```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

#### Read the data file

```python
catch=pd.read_csv('EGOALLO_PROTOCOL_OUTCOMES_AND_PERFORMANCES.csv')
catch.head()
```

```python
def keep_learning(group):
    
    '''
    Keep pre-catch sessions only and add first choice column
    '''
    group_=group.copy()
    group_['first_choice']=group_.loc[group_.catch_nr==1, 'outcome'].iloc[0]
    catch_index=group_[group_.catch_nr==1].index.tolist()[0]
    return group_.loc[:catch_index]
```

```python
sessions_to_catch=catch.groupby(['rat']).apply(keep_learning).reset_index(drop=True)
```

```python
sessions_to_catch.head(30)
```

```python
sessions_to_catch.outcome.value_counts()
```

```python
sns.set(style='white', context='talk')
plt.figure(dpi=300)
g=sns.relplot(
    kind='line',
    data=sessions_to_catch,
    y='performance',
    x='session_nr',
    col='first_choice', marker='o', sizes=.01,
    hue='rat', height=3.5, aspect=1.3, 
    palette=['black']*12, legend=False,
    )
g.set(xlabel='Session', xticks=range(0,8,2), ylabel='Performance(%)')
titles=['Action on first catch', 'Location on first catch']
for ax, title in zip(g.axes.flatten(), titles):
    ax.axhline(80, color='black', linestyle='dashed', linewidth=1)
    ax.set(title=title)
```

```python
first_catch = catch[catch.catch_nr==1]
first_catch.head(3)
```

```python
sns.set(style='white', context='talk')
plt.figure(dpi=300, figsize=(3.5,2.5))
g=sns.barplot(
    data=first_catch,
    y='session_nr',
    x='outcome',
    palette=['mediumseagreen','royalblue'],
    capsize=.2, ci='sd', errwidth=1.5)

g.set(ylabel='Session number', xlabel='First non-match catch',
     yticks=range(0,8,2), xticks=[0,1], xticklabels=['action', 'location'])
sns.despine()
```

#### Normality testing

```python
from scipy.stats import shapiro
first_catch.groupby(['outcome']).apply(shapiro)
```

#### Significance test

```python
from scipy.stats import mannwhitneyu
mannwhitneyu(first_catch.loc[first_catch.outcome=='A', 'session_nr'],
            first_catch.loc[first_catch.outcome=='E', 'session_nr'])
```
